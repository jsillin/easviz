```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gfr_logger.csv')
```

```python
df
```

```python
tmax = df['tmax']
```

```python
tmin = df['tmin']
rh = df['rhum']
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(211)
ax.plot(tmax,color='red')
ax.plot(tmin,color='blue')
ax.plot(rh,color='green')
ax.set_title('Observed Weather at Game Farm Rd')
ax.set_ylim(0,100)

ax1 = fig.add_subplot(212)
```
